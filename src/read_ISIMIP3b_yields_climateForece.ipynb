```python
variables=['tas','pr']
models=['gfdl-esm4','ipsl-cm6a-lr','mpi-esm1-2-hr','mri-esm2-0','ukesm1-0-ll']
periods=['2015_2020',
         '2021_2030',
         '2031_2040',
         '2041_2050',
         '2051_2060',         
         '2061_2070',
         '2071_2080',  
         '2081_2090',
         '2091_2100',         
        ]
ssp=['ssp126','ssp585']
```

```python
for i in range(5):
    for j in range(2):
        for k in range(2):
#           for kk in range(9):
            fnstr=Path+models[i]+'_r1i1p1f2_w5e5_'+ssp[j]+'_'+variables[k]+'_global_daily_'+'*.nc'
            fns=glob.glob(fnstr)
            print((models[i],ssp[j],variables[k],len(fns)))
```

```python
# 94 years from 2006 to 2099; 4 models and 2 rcps
annual_pr=np.empty(shape=(85,5,2))
annual_tm=np.empty(shape=(85,5,2))
```

```python
imodel=0
for model in models:
    ircp=0
    print(model)
    for rcp in ssp:
        print(rcp)
        iperiod=0
        for period in periods:
            if imodel==4:
                file=model+'_r1i1p1f2_w5e5_'+rcp+'_'+'pr_global_daily_'+period+'.nc'
            else:
                file=model+'_r1i1p1f1_w5e5_'+rcp+'_'+'pr_global_daily_'+period+'.nc'
            fullpath=Path+file
            ds=xr.open_dataset(fullpath)
            cropped_ds = ds.sel(lat=slice(max_lat,min_lat), lon=slice(min_lon,max_lon))
#average over the region to get time series
            pr=cropped_ds['pr'].mean(dim=('lat','lon'))*3600*24
#                 if not os.path.isfile(fullpath):
#                     print(fullpath)
            if iperiod==0:
                data=pr
                iperiod=1
            else:
                data=xr.concat([data,pr],dim='time')
        annualpr=data.groupby('time.year').sum().values
        annual_pr[:,imodel,ircp]=annualpr[:85]
        ircp=ircp+1
    imodel=imodel+1
```
